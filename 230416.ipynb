```python
#https://school.programmers.co.kr/learn/courses/30/lessons/178871
#얀에서는 매년 달리기 경주가 열립니다. 
#해설진들은 선수들이 자기 바로 앞의 선수를 추월할 때 추월한 선수의 이름을 부릅니다. 
#예를 들어 1등부터 3등까지 "mumu", "soe", "poe" 선수들이 순서대로 달리고 있을 때, 
#해설진이 "soe"선수를 불렀다면 2등인 "soe" 선수가 1등인 "mumu" 선수를 추월했다는 것입니다. 
#즉 "soe" 선수가 1등, "mumu" 선수가 2등으로 바뀝니다.

#선수들의 이름이 1등부터 현재 등수 순서대로 담긴 문자열 배열 players와 해설진이 부른 이름을 담은 문자열 배열 callings가 매개변수로 주어질 때, 
#경주가 끝났을 때 선수들의 이름을 1등부터 등수 순서대로 배열에 담아 return 하는 solution 함수를 완성해주세요.

def make_dict(players): #리스트를 딕셔너리로 최적화 하자
    players_dict = {}
    for i, player in enumerate(players):
        players_dict[player] = i #플레이어 이름 : 플레이어의 현재 위치
    return players_dict

def change_rank(players_dict, players, player): #추월 함수
    player_index = players_dict[player] # 받은 딕셔너리에서 플레이어의 현재 위치를 불러옴
    #리스트에 추월 구현(리스트의 이전 플레이어 위치에 추월하는 플레이어 이름으로 변경, 추월하는 플레이어 위치에 이전 플레이어 이름으로 변경)
    players[player_index-1], players[player_index] =  players[player_index], players[player_index-1] 
    #딕셔너리에 변경된 이전 플레이어 위치와 추월 플레이어 값 갱신
    players_dict[players[player_index]] = player_index #추월 당한 플레이어
    players_dict[players[player_index-1]] = player_index-1 #추월한 플레이어

def solution(players, callings):
    players_dict = make_dict(players) #플레이어 위치 테이블 생성
    for player in callings: #calling에서 이름이 불릴 때마다 순차 진행
        change_rank(players_dict, players, player) #서로의 위치를 변경
    return players

print(solution(["mumu", "soe", "poe", "kai", "mine"], ["kai", "kai", "mine", "mine"])) #answer ['mumu', 'kai', 'mine', 'soe', 'poe']
#for문 중첩 반복은 time complexity가 기하급수적으로 높아질 우려가 있으므로 dict로 최적화한다.
#리스트의 값 변경 뿐만 아니라 딕셔너리도 갱신해 주어야 한다.
```
