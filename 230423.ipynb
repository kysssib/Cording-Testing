## **문제 설명**
x축과 y축으로 이루어진 2차원 직교 좌표계에 중심이 원점인 서로 다른 크기의 원이 두 개 주어집니다. 반지름을 나타내는 두 정수 `r1`, `r2`가 매개변수로 주어질 때, 두 원 사이의 공간에 x좌표와 y좌표가 모두 정수인 점의 개수를 return하도록 solution 함수를 완성해주세요.
※ 각 원 위의 점도 포함하여 셉니다.

---

### **제한 사항**
1 ≤ `r1` < `r2` ≤ 1,000,000

---

#### **입출력 예**
|r1|r2|result|
|--|--|--|
|2|3|20|

---

```
import math #정확성을 위해서 math 불러옴

def solution (r1,r2):
    answer = 0
    for i in range(1, r2 + 1): # 점의 끝은 원 길이보다 하나 더 +1 1부터 시작은 4배할때 겹치는거 제거 
        y2 = r2 ** 2 - i ** 2 #피타고라스 삼각형의 삼각형 높이 값
        if i < r1: y1 = r1 ** 2 - i ** 2
        else: y1 = 0 #작은 원 반지름 길이보다 더 길면 안됨
        y2_points = math.floor(math.sqrt(y2)) #y2의 길이를 구해 정수형으로 변경 위해서 버림
        y1_points = math.ceil(math.sqrt(y1)) #y2의 길이를 정수형으로 변경 위해서 올림(둘 사이의 거리를 구하기 위함)
    
        answer += (y2_points - y1_points + 1) * 4 #(둘 사이의 거리 = y2 - y1 +1)에 4배 해야함
    return answer

print(solution(2,3))
```

```
#하려던거 -> 오류 남
import math
def distance_2_points(x1, y1, x2, y2):
    distance_x = x2 - x1
    distance_y = y2 - y1
    distance_x *= distance_x
    distance_y *= distance_y
    return math.sqrt(distance_x + distance_y)

def count_sub_point(radius):
    count_points = 0
    same_point = 0
    same_radius = 0
    for i in range(0, radius+1):
        for j in range(radius,-1,-1):
            distance = distance_2_points(0,0,i,j)
            if distance > radius:
                if i == j :
                    same_point += 1
                    break
                count_points += 1
            elif distance == radius:
                same_radius += 1            
            else:
                break
    
    return count_points * 2 + same_point, same_radius


def in_circle_point(radius):
    square = radius * 2 + 1
    square *= square
    qurter, same_radius = count_sub_point(radius)
    return square - (qurter * 4), same_radius * 4

def solution(r1, r2):
    r1_points, r1_outline_points = in_circle_point(r1)
    r2_points, r2_outline_points = in_circle_point(r2)
    answer = r2_points - r1_points + r1_outline_points
    return answer
```
