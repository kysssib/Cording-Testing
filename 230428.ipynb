```python
def solution(cards1, cards2, goal):
    answer = 'Yes'
    deck = {}
    for i, card1 in enumerate(cards1):
        deck[card1] = i+1
    for i, card2 in enumerate(cards2):
        deck[card2] = i+111
    
    deck_collection = [[],[]]

    before_word = goal[0]
    deck_num = 0 if deck[before_word] <= 10 else 1

    for word in goal[1:]:
        now_val = deck[word]
        before_val = deck[before_word]
        #print(before_word, before_val)
        #print(word, now_val)
        if abs(now_val - before_val) <= 100:
            pass
        else:
            deck_num = (deck_num + 1) % 2
        before_same_deck = deck_collection[deck_num]
        if before_same_deck != []:
            if now_val - before_same_deck[-1] != 1:
                answer = 'No'
                return answer
        
        deck_collection[deck_num].append(now_val)
        before_word = word
    return answer

print(solution(["i", "drink", "water"], ["want", "to"], ["i", "want", "to", "drink", "water"]))
print(solution(["i", "water", "drink"], ["want", "to"], ["i", "want", "to", "drink", "water"]))
```
